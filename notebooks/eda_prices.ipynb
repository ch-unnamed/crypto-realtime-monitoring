EDA para precios de criptomonedas
Análisis de la serie temporal de precios de BTCUSDT obtenida en tiempo real desde la API de Binance y almacenada en PostgresSQL a través del pipeline (producer-> Kafka -> consumer). Se realiza un análisis exploratorio básico y se generan features como retornos y volatilidad rolling.

```python
#Imports y conexión a BD en Localhost
import pandas as pd 
import sqlalchemy as sa
from dotenv import load_dotenv
import os
import matplotlib.pyplot as plt
import seaborn as sns

load_dotenv()

DB_USER = os.getenv("DB_USER")
DB_PASS = os.getenv("DB_PASSWORD")
DB_HOST = "localhost"
DB_PORT = os.getenv("DB_PORT")
DB_NAME = os.getenv("DB_NAME")


engine = sa.create_engine(
    f"postgresql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)

query = """ SELECT id, symbol, price, volume, ts_event, source
FROM raw_prices
WHERE symbol = 'BTCUSDT' 
ORDER BY ts_event
LIMIT 2000; """

df = pd.read_sql(query, engine)
df.head()
df.info()
df.describe()
df.isna().mean()

df["ts_event"] = pd.to_datetime(df["ts_event"])
df = df.sort_values("ts_event")

df = df.set_index("ts_event")

df.head()
```
